```python
import pandas as pd
```

```python
c=pd.read_csv("tmdb_5000_credits.csv")
m=pd.read_csv("tmdb_5000_movies.csv")
```

```python
c.head(1)
```

```python
m.head(1)
```

```python
data=pd.merge(c,m,on="title")
```

```python
data.head(1)
```

```python
dataset=data[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
dataset.head(1)
```

```python
dataset["genres"][0]
```

```python
import ast
```

```python
def genres(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i["name"])
    return l
```

```python
dataset.loc[:,"genres"]=dataset["genres"].apply(genres)
```

```python
dataset["keywords"][0]
```

```python
dataset.loc[:,"keywords"]=dataset["keywords"].apply(genres)
```

```python
dataset.head(1)
```

```python
dataset["cast"][0]
```

```python
def cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i["name"])
            counter+=1
    return l
```

```python
dataset.loc[:,"cast"]=dataset["cast"].apply(cast)
```

```python
dataset.head(1)
```

```python
dataset["crew"][0]
```

```python
def crew(text):
    l=[]
    for i in ast.literal_eval(text):
        if i.get("job")=="Director":
            l.append(i["name"])
    return l
```

```python
dataset.loc[:,"crew"]=dataset["crew"].apply(crew)
```

```python
dataset.head(1)
```

```python
type(dataset.iloc[0]["keywords"])
```

```python
def space(text):
    l=[]
    for i in text:
        a=i.replace(" ","")
        l.append(a)
    return l
```

```python
dataset.loc[:,"genres"]=dataset["genres"].apply(space)
```

```python
dataset.loc[:,"keywords"]=dataset["keywords"].apply(space)
```

```python
dataset.head(1)
```

```python
dataset.loc[:,"cast"]=dataset["cast"].apply(space)
dataset.loc[:,"crew"]=dataset["crew"].apply(space)
```

```python
dataset.head(1)
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.dropna(subset="overview")
```

```python
dataset.isnull().sum()
```

```python
dataset.loc[:,"overview"]=dataset["overview"].apply(lambda x:x.split())
```

```python
dataset.head(1)
```

```python
dataset["tags"]=dataset["overview"]+ dataset["genres"]+ dataset["keywords"]+ dataset["cast"]+ dataset["crew"]
```

```python
dataset.head(1)
```

```python
dataset["tags"][0]
```

```python
final_data=dataset[["movie_id","title","tags"]]
```

```python
final_data.head(1)
```

```python
final_data.loc[:,"title"]=final_data["title"].apply(lambda x: x.lower())
```

```python
final_data.loc[:,"tags"]=final_data["tags"].apply(lambda x:" ".join(x))
```

```python
final_data.head(1)
```

```python
final_data.loc[:,"tags"]=final_data["tags"].apply(lambda x:x.lower())
```

```python
final_data.head(1)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stemming(text):
    l=[]
    for i in text.split():
        a=ps.stem(i)
        l.append(a)
    return " ".join(l)
```

```python
final_data.loc[:,"tags"]=final_data["tags"].apply(stemming)
```

```python
final_data.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(final_data["tags"]).toarray()
```

```python
print(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
index=final_data[final_data["title"]=="titanic"].index[0]
```

```python
similarity[index]
```

```python
distance=similarity[index]
```

```python
list(enumerate(distance))
```

```python
sorted(list(enumerate(distance)),reverse=True)
```

```python
sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])
```

```python
sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[0:6]
```

```python
title=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[0:6]
```

```python
for i in title:
    print(final_data.iloc[i[0]]["title"])
```

```python
def Movie_Recommender(movie):
    try:
        if movie in final_data["title"].values:
            index=final_data[final_data["title"]==movie].index[0]
            distance=similarity[index]
            title=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

            for i in title:
                print(final_data.iloc[i[0]]["title"])
    
        else:
            print("Movie Not Found in the Database !")

    except Exception as e:
        print("Movie not found !")
```
